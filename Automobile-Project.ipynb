```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
auto = pd.read_csv('Automobile_data.csv.csv')
```

```python
auto.head()
```

```python
sns.distplot(auto['normalized-losses'])
plt.show
```

```python
sns.distplot(auto['engine-size'])
plt.show
```

```python
sns.distplot(auto['city-mpg'], kde=False, rug=True);
plt.show
```

```python
sns.distplot(auto['engine-size'], kde=False, rug=True)
plt.show
```

```python
sns.jointplot(auto['engine-size'], auto['horsepower'])
plt.show()
```

```python
sns.jointplot(auto['engine-size'], auto['horsepower'], kind="hex")
plt.show()
```

```python
sns.jointplot(auto['engine-size'], auto['horsepower'], kind="reg")
plt.show()
```

```python
sns.jointplot(auto['engine-size'], auto['horsepower'], kind="kde")
plt.show()
```

```python
sns.pairplot(auto[['normalized-losses', 'engine-size', 'horsepower']])
```

```python
sns.stripplot(auto['fuel-type'], auto['horsepower'])
```

```python
sns.stripplot(auto['fuel-type'], auto['horsepower'], jitter = True)
```

```python
sns.swarmplot(auto['fuel-type'], auto['horsepower'])
```

```python
sns.boxplot(auto['fuel-type'], auto['horsepower'])
```

```python
sns.barplot(auto['body-style'], auto['horsepower'])
```
